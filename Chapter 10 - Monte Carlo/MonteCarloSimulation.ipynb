```python
import numpy as np
from numba import jit 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
sns.set_theme(
    style = "darkgrid",
    palette = "bright",
)

plt.rc('figure', figsize = (10, 5))
```

## <center> Importance Sampling and Statistical Mechanics </center> 

## <center> $ P(E_{i}) = \frac{e^{\beta E_{i}}}{Z} $ </center>

## <center> $ Z = \sum_{i=1}^{N} e^{-\beta E_{i}} $ </center>

```python
@jit(nopython = True)
def compute_boltzmann_formula(beta, e, E):
    """
    
    """
    
    Z = np.sum(np.exp(-beta * E))
    P = np.exp(-beta * e) / Z
    
    return P
```

```python
E = np.linspace(0, 1, 1000)
boltzmann = np.copy(E)
beta = 5

for i, e in enumerate(E):
    boltzmann[i] = compute_boltzmann_formula(beta, e, E)

ax = sns.lineplot(boltzmann)

ax.set(
    title = "Boltzmann Formula",
    xlabel = "Energy",
    ylabel = "Probability"
)

plt.show()
```
