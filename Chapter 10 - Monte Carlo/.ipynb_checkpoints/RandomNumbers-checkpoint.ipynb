```python
import numpy as np
from numba import jit 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
sns.set_theme(
    style = "darkgrid",
    palette = "bright",
)

plt.rc('figure', figsize = (10, 5))
```

# <center> Global Variable Initialization </center> 

```python
# Sample Size
N = 10000

# X Axis
X = np.arange(0, N, 1)
```

## <center> Linear Congruential Random Number Generator </center>

## <center> $ x' = (a \cdot x + c) \cdot mod(m) $ </center>

```python
@jit(nopython = True, parallel = True)
def generate_random_numbers(seed):
    
    # Initialize Constants
    a = 1664525
    c = 1013904223
    m = 4294967296
    
    # Return this
    arr = np.zeros(N)
    
    # Initial Procedure
    arr[0] = (a * seed + c) % m
    
    for i in range(1, N):
        arr[i] = (a * arr[i - 1] + c) % m
    
    return arr
```

```python
rv = generate_random_numbers(seed = 1)

plt.scatter(X, rv)
plt.title("Psuedorandom Numbers")
plt.show()
```

```python
rv_uniform = rv / 4294967296

plt.scatter(X, rv_uniform)
plt.title("Uniform Psuedorandom Numbers")
plt.show()
```

## <center> Exponentially Distributed Values </center> 

## <center> $ x = -\frac{1}{\mu} \cdot ln(1 - z)$ </center>

```python
@jit(nopython = True, parallel = True)
def generate_exponential():
    
    # Initialize Constants
    mu = 1
    
    # Return This
    arr = np.zeros(N)
    
    # Generate Uniform Random Variables
    rv_uniform = generate_random_numbers(seed = 1) / 4294967296
    
    for i in range(N):
        arr[i] = -(1 / mu) * np.log(1 - rv_uniform[i])
    
    return arr
```

```python
rv_exponential = generate_exponential()

plt.hist(rv_exponential, bins = 100, density = True)
plt.title("Psuedorandom Exponential Variables")
plt.xlabel("X")
plt.ylabel("Frequency")
plt.show()
```

## <center> Box Muller Transform </center>

## <center> $ R = \sqrt{-2 \cdot ln(U_{1})} $ </center>

## <center> $ Z_{0} = R \cdot cos(2 \cdot \pi \cdot U_{2}) $ </center>

## <center> $ Z_{1} = R \cdot sin(2 \cdot \pi \cdot U_{2}) $ </center>

```python
@jit(nopython = True, parallel = True)
def box_muller_transform():
    
    # Generate Uniform Random Variables
    rv_uniform1 = generate_random_numbers(seed = 1) / 4294967296
    rv_uniform2 = generate_random_numbers(seed = 2) / 4294967296

    # Return This
    arr = []
    
    for i in range(N):
        
        R = np.sqrt(-2 * np.log(rv_uniform1[i]))
        arr.append(R * np.cos(2 * np.pi * rv_uniform2[i]))
        arr.append(R * np.sin(2 * np.pi * rv_uniform2[i]))

    return np.array(arr)
```

```python
rv_gaussian = box_muller_transform()

plt.hist(rv_gaussian, bins = 100, density = True)
plt.title("Psuedorandom Gaussian Variables")
plt.xlabel("X")
plt.ylabel("Frequency")
plt.show()
```
