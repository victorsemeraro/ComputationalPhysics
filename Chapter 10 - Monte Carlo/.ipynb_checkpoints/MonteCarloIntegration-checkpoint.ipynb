## <center> $ I = \int_{0}^{2} sin^2 \cdot [\frac{1}{x \cdot (2 - x)}] dx $ </center>

## <center> Actual Value = 1.4522 </center>

```python
@jit(nopython = True)
def f(x):
    
    return (np.sin(1 / (x * (2 - x))))**2
```

```python
x = np.linspace(0.01, 1.99, 1000)
y = f(x)

plt.plot(x, y)
plt.title("Complicated Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## <center> Simple Monte Carlo Integration </center>

## <center> $ I \approx \frac{k \cdot A}{N} $ </center>

```python
@jit(nopython = True, parallel = True)
def monte_carlo_integration():
    
    # Sample Size
    N = 1000
    
    # Generate Random Numbers
    x = np.random.uniform(0, 2, N)
    y = np.random.uniform(0, 1, N)
    
    # Compute Function Values
    fx = f(x)
    
    # Initialize Count
    count = 0
    
    for i in range(N):
        
        if y[i] < fx[i]:
            count += 1
            
    I = 2 * count / N
    
    return I
```

```python
I = monte_carlo_integration()

print("Monte Carlo Integral: ", I)
```

## <center> Mean Value Method Derivation </center>

## <center> $ I = \int_{a}^{b} f(x) dx $ </center> 

## <center> $ f_{avg} = \frac{1}{b - a} \int_{a}^{b} f(x) dx = \frac{I}{b - a}$ </center> 

## <center> $ I = (b - a) \cdot f_{avg} $ </center> 

## <center> $ I \approx \frac{b - a}{N} \sum_{i=1}^{N} f(x_{i}) $ </center>

```python
@jit(nopython = True, parallel = True)
def mean_value_integration():
    
    # Sample Size
    N = 1000
    
    # Integral Bounds
    a = 0
    b = 2
    
    # Generate Random Numbers
    x = np.random.uniform(0, 2, N)
    
    # Compute Function Values
    fx = f(x)
    
    # Initialize Count
    total = 0
    
    for i in range(N):
        total += fx[i] 
            
    I = ((b - a) / N) * total
    
    return I
```

```python
I = mean_value_integration()

print("Mean Value Integral: ", I)
```

## <center> Multidimensional Integration </center> 

## <center> Importance Sampling Method Derivation </center>

## <center> $ g_{avg} = \frac{\int_{a}^{b} w(x) g(x) dx}{\int_{a}^{b} w(x) dx} $ </center> 

## <center> $ I = \int_{a}^{b} f(x) dx $ </center>

## <center> $ g_{avg} = \frac{f(x)}{w(x)} $ </center> 

## <center> $ g_{avg} = \frac{\int_{a}^{b} w(x) \frac{f(x)}{w(x)} dx}{\int_{a}^{b} w(x) dx} $ </center> 

## <center> $ g_{avg} = \frac{I}{\int_{a}^{b} w(x) dx} $ </center> 

## <center> $ I = g_{avg} \int_{a}^{b} w(x) dx $ </center>

## <center> $ I \approx \frac{1}{N} \sum_{i=1}^{N} \frac{f(x_{i})}{w(x_{i})} \int_{a}^{b} w(x) dx $ </center>

```python
@jit(nopython = True)
def weight_function(x):
    """
    
    """
        
    return np.exp(-x)
```

```python
@jit(nopython = True)
def weight_function_integral(a, b):
    """
    
    """
    
    return -np.exp(-b) + np.exp(-a)
```

```python
@jit(nopython = True, parallel = True)
def importance_sampling_integration():
    """
    
    """
    
    # Sample Size
    N = 1000
    
    # Integral Bounds
    a = 0
    b = 2
    
    # Generate Random Numbers
    x = np.random.uniform(0, 2, N)
    
    # Compute Function Values
    fx = f(x)
    wx = weight_function(x)
    
    # Initialize Count
    total = 0
    
    for i in range(N):
        total += (fx[i] / wx[i]) 
            
    I = (1 / N) * total * weight_function_integral(a, b)
    
    return I
```

```python
I = importance_sampling_integration()

print("Importance Sampling Integral: ", I)
```
