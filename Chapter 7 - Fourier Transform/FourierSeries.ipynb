# <center> Fourier Series </center> 

```python
import numpy as np
from numba import jit 
import matplotlib.pyplot as plt 
```

## <center> Initialize Global Variables </center> 

```python
N = 1000
x = np.linspace(0, 2, N)
k = np.arange(0, N, 1)
```

## <center> </center>

## <center> $ f(x) = \sum_{k=0}^{\infty} \alpha_{k} \cdot cos(\frac{2 \pi k x}{L}) $ </center>

```python
@jit(nopython = True, parallel = True)
def symmetric_fourier_series(alpha, L):
    
    # Return This
    arr = np.zeros(N)
    
    # Initialize Local Variables
    running_sum = 0 
    
    for i in range(N):
        arr[i] = alpha * np.cos(2 * np.pi * k[i] * x[i] / L)
    
    return arr
```

```python
symmetric_fourier = symmetric_fourier_series(alpha = 2, L = 0.5)

plt.plot(x, symmetric_fourier)
plt.title("Symmetric Fourier Series")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## <center> Anti Symmetric Fourier Series </center> 

## <center> $ f(x) = \sum_{k=1}^{\infty} \beta_{k} \cdot cos(\frac{2 \pi k x}{L}) $ </center>

```python
@jit(nopython = True, parallel = True)
def anti_symmetric_fourier_series(beta, L):
    
    # Return This
    arr = np.zeros(N)
    
    # Initialize Local Variables
    running_sum = 0 
    
    for i in range(N):
        arr[i] = beta * np.sin(2 * np.pi * k[i] * x[i] / L)
    
    return arr
```

```python
anti_symmetric_fourier = anti_symmetric_fourier_series(beta = 2, L = 0.5)

plt.plot(x, anti_symmetric_fourier)
plt.title("Anti Symmetric Fourier Series")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
